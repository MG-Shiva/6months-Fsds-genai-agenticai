```python
# bi-variaant => plotting the graph using two  varibles
# boxplot => draws a box plot to show distributions with respect to categories.

vis4 = sns.boxplot(data =df, x="IncomeGroup", y="BirthRate")
plt.show(vis4)
```

```python
# In statistics => An outlier in a graph is a data point that significantly deviates from the 
# --overall pattern or trend of the other data points

# lines are min-max(range)
# circles => outliers ( outliers will impact all ml & dl algorithms
# outlier is called as ANAMOLY DETECTION
# line inside box => Mean / median / average of distribution
```

```python

# ML - 2 algorithm which handle outlier
  # logistic regression algorithm (SIGMOID) probability
  # naiven bayes algorithm (BAYSIAN THEOREM)
```

```python
# Plot data and regression model fits across a FacetGrid.
# A FacetGrid in Seaborn is a multi-plot grid for visualizing data, particularly useful when you want
# --to examine the distribution of one or more variables across different subsets of your dataset.

vis5 = sns.lmplot(data = df, x ='InternetUsers', y = 'BirthRate')
plt.show(vis5)
```

```python
# line => Regression line

# A regression line is a straight line that describes how a response variable y changes
   # --as an explanatory variable x changes.

#  A regression line can be used to predict the value of y for a given value of x
```

```python
# removing the regression from the plot

vis6 = sns.lmplot(data = df, x ='InternetUsers', y = 'BirthRate', fit_reg=False)
plt.show(vis6)
```

```python
# hue for colour variation

vis7 = sns.lmplot(data = df, x ='InternetUsers', y = 'BirthRate', fit_reg=False, hue='IncomeGroup')
plt.show(vis7)
```
